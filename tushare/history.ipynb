```python
#将绘画框进行对象化
fig=plt.figure(figsize=(24,8), dpi=80)

#将p1定义为绘画框的子图，211表示将绘画框划分为2行1列，最后的1表示第一幅图
p1=fig.add_subplot(111)
p1.bar(df_m.index,df_m['close'])

#将p2定义为绘画框的子图，211表示将绘画框划分为2行1列，最后的2表示第二幅图
p2=fig.add_subplot(211)
p2.scatter(df_m.index,df_m['open'])

plt.show()
```
